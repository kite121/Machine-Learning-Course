# Lab 11 - Decision Tree. Self practice.

We will implement Decision Tree algorithm for the regression problem.

Let's prepare the dataset.

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

# Create a random dataset
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))
```

```python
X[:10]
```

```python
y[:10]
```

Let's create a Decision Tree model for the regression problem.

```python
# Fit regression model
# Create 2 Decision Tree Regressor models with 2 and 5 max_depths
regr_1 = DecisionTreeRegressor(max_depth=2, random_state=0)
regr_2 = DecisionTreeRegressor(max_depth=5, random_state=0)

# Train both models on the dataset
regr_1.fit(X, y)
regr_2.fit(X, y)
```

Once our models are trained, we can use them for prediction.

```python

X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]

# Predict
# write 2 lines of code for the prediction on the X_test by each of the created models
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
```

Let's plot the results and see how different depths affext the model performance

```python

# Plot the results
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black",
            c="darkorange", label="data")
plt.plot(X_test, y_1, color="cornflowerblue",
         label="max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```
